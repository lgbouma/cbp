```python
foo = np.random.uniform(low=np.log(5), high=np.log(365), size=1e4)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ax = sns.distplot(foo, kde=False)
ax.set(xlabel='logP');
```

```python
ax = sns.distplot(np.e**foo, kde=False, norm_hist=True)
ax.set(xlabel='e^logP = P')
P1, P2 = 5, 365
P = np.arange(5,365,0.1)
ax.plot(P, 1/(np.log(P2) - np.log(P1))/P, 'k-')
ax.set(title='histogram numerics. black line analytics');
```

```python
ln_Rp = np.random.uniform(low=np.log(0.04), high=np.log(0.2), size=1e4)
Rp = np.e**ln_Rp
δ = Rp**2
ax = sns.distplot(δ, kde=False, norm_hist=True)
ax.set(xlabel='transit depth (relative flux units)');
```

```python
ln_RpbyRs = np.random.uniform(low=np.log(0.01), high=np.log(0.1), size=1e4)
RpbyRs = np.e**ln_RpbyRs
δ = RpbyRs**2
ax = sns.distplot(δ, kde=False, norm_hist=True)
ax.set(xlabel='transit depth (relative flux units)',
      ylabel='number (pdf, normalized)');
```

```python
ax = sns.distplot((δ**(1/2.)*(1.5*u.R_sun)).to(u.Rearth), kde=False, norm_hist=True)
ax.set(xlabel='assumed Rp [R_earth]', ylabel='pdf');
```

```python
ln_RpbyRs = np.random.uniform(low=np.log(0.01), high=np.log(0.1), size=1e4)
RpbyRs = np.e**ln_RpbyRs
δ = RpbyRs**2
ax = sns.distplot(δ, kde=False, norm_hist=True,
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
ax.set(xlabel='transit depth (relative flux units)',
      ylabel='cdf');
```
