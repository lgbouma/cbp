```python
import numpy as np
```

```python
from scipy.interpolate import interp1d
```

```python
degs = np.array([2,5,10,15])
pts = np.array([1e2,4.3e2,3e3,4e3])
fn = interp1d(pts, degs, kind='linear', 
             bounds_error=False, 
             fill_value=(2, 15))
npts = np.arange(1,1e4,1)
legendredegs = np.floor(fn(npts))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
f, ax = plt.subplots()
ax.plot(npts, legendredegs, 'k-')
ax.set(xscale='log', ylim=[0,17],
      xlabel='number of points in quarter',
      ylabel='finite legendre series degree\n(legendredeg)')
f.savefig('plots/legendredeg_vs_npts_per_quarter.pdf')
f.show()
```
