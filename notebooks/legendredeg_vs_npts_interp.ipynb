```python
import numpy as np
```

```python
from scipy.interpolate import interp1d
```

```python
degs = np.array([4,5,6,10,15])
pts = np.array([1e2,3e2,5e2,1e3,3e3])
fn = interp1d(pts, degs, kind='linear', bounds_error=False, fill_value=(min(degs), max(degs)))
npts = np.arange(1,1e4,1)
legendredegs = list(map(int,np.floor(fn(npts))))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
f, ax = plt.subplots()
ax.plot(npts, legendredegs, 'k-')
ax.set(xscale='log', ylim=[0,20],
      xlabel='number of points in timegroup',
      ylabel='finite legendre series degree\n(legendredeg)')
f.savefig('plots/legendredeg_vs_npts_per_timegroup_ctd.pdf')
f.show()
```
