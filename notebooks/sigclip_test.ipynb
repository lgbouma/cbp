```python
from astrobase import lcmath
import matplotlib.pyplot as plt
import batman
```

```python
import numpy as np
```

```python
params = batman.TransitParams()
params.t0 = 0.                       #time of inferior conjunction
params.per = 1.                      #orbital period
params.rp = 0.1                      #planet radius (in units of stellar radii)
params.a = 15.                       #semi-major axis (in units of stellar radii)
params.inc = 87.                     #orbital inclination (in degrees)
params.ecc = 0.                      #eccentricity
params.w = 90.                       #longitude of periastron (in degrees)
params.u = [0.1, 0.3]                #limb darkening coefficients
params.limb_dark = "quadratic"       #limb darkening model
```

```python
t = np.linspace(-1.5, 1.5, 1e4)
m = batman.TransitModel(params, t)    #initializes model

flux_m = m.light_curve(params)          #calculates light curve

flux_err = 0.05*np.random.random(size=len(flux_m))

flux = flux_m + flux_err
```

```python
%matplotlib inline

plt.close('all')
plt.plot(t, flux, 'k-o', ms=0.5, lw=0.1)
plt.xlabel("Time from central transit")
plt.ylabel("Relative flux")
plt.show()
```

```python
#errs are irrelevant for now
times, flux, errs = t, flux, np.zeros_like(t)

stime, sflux, serrs = lcmath.sigmaclip_lc(times, flux, errs, 
                             isflux=True, sigclip=[3, 0.1])

plt.close('all')
plt.plot(stime, sflux, 'k-o', ms=0.5, lw=0.1)
plt.xlabel("Time from central transit")
plt.ylabel("Relative flux (sigclipped)")
plt.show()
```

```python
#now check the mags functionality

mags = -5/2. * np.log10(flux)

plt.close('all')
plt.plot(t, mags, 'k-o', ms=0.5, lw=0.1)
plt.xlabel("Time from central transit")
plt.ylabel("Relative magnitude")
plt.show()
```

```python
times = t

stimes, smags, serrs = lcmath.sigmaclip_lc(times, mags, errs,
                              isflux=False, sigclip=[3, 0.1])

plt.close('all')
plt.plot(stimes, smags, 'k-o', ms=0.5, lw=0.1)
plt.xlabel("Time from central transit")
plt.ylabel("Relative magnitude (sigclipped)")
plt.show()
```
