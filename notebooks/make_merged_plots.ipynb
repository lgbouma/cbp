```python
import os
import numpy as np, matplotlib.pyplot as plt
from IPython.display import Image
import matplotlib.image as mpimg
from astropy.io import ascii
```

```python
ids = np.array([f[:-4] for f in os.listdir('../data/keplerebcat_LCs/data/raw/')])

#get Kepler EB data (e.g., the period)
keb_path = '../data/kepler_eb_catalog_v3.csv'
cols = 'KIC,period,period_err,bjd0,bjd0_err,morph,GLon,GLat,kmag,Teff,SC'
cols = tuple(cols.split(','))
tab = ascii.read(keb_path)
currentcols = tab.colnames
for ix, col in enumerate(cols):
    tab.rename_column(currentcols[ix], col)
tab.remove_column('col12') # now table has correct features


#get image data and make plots
ptypes = ['raw','dtr','etv']

for ebid in ids:
    plt.close('all')
    outpath = '../results/keplerebcat_diagnostics/{:s}.png'.format(ebid)

    if not os.path.exists(outpath) and str(ebid) not in ['2970804','8197406','5983351','2856960']:
        imgpaths = []
        for ptype in ptypes:
            imgpaths.append('../data/keplerebcat_LCs/plots/{:s}/{:s}.{:s}.png'.format(ptype,ebid,ptype))
        #now that we have image paths for this EB, get period, morph, mag, then image data

        f, axs = plt.subplots(figsize=(8*len(imgpaths),6), 
                              nrows=1, ncols=len(imgpaths), dpi=200)
        axs = axs.flatten()

        for ix, imgpath in enumerate(imgpaths):
            ax = axs[ix]
            print(imgpath)
            img=mpimg.imread(imgpath)
            ax.imshow(img)
            ax.axis('off')

        f.text(0.3,0.99,'KIC ID: {:s},   period: {:.4g} d,   morph: {:.4g},   kmag: {:.4g}'.\
               format(ebid,
                      float(tab[tab['KIC']==int(ebid)]['period']),
                      float(tab[tab['KIC']==int(ebid)]['morph']),
                      float(tab[tab['KIC']==int(ebid)]['kmag'])), 
               fontsize='xx-large')

        f.tight_layout(w_pad=-13.5, pad=-2)

        f.savefig(outpath, bbox_inches='tight', w_pad=-13.5, pad=-2)    
```
