Compare the distributions of EB periods from my sample with Klagyivik et al 2016

```python
import pandas as pd, numpy as np
from astropy.io import ascii
```

```python
tab = ascii.read('../data/Klagyivik_Deeg_et_al_2016_table1.dat')
df = tab.to_pandas()
df = df.rename(index=str, columns={'col1':'corot_id', 'col2':'period'})

% matplotlib inline
import seaborn as sns

ax = sns.distplot(np.log10(df['period']), kde=False)
ax.set_xlabel('log10(period)')
ax.set_ylabel('number');
print(len(df[df['period']<3]))
```

```python
df = pd.read_csv('../data/kebc_v3_170611_headerless.csv')
ax = sns.distplot(np.log10(df['period']), kde=False)
ax.set_xlabel('log10(period)')
ax.set_ylabel('number');
print(len(df[df['period']<3]))
```
